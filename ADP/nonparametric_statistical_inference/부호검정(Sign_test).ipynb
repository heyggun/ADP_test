## `부호검정(Sign test)`

http://contents.kocw.net/KOCW/document/2013/koreasejong/HongSungsik4/13.pdf

```python
iq_lst = [128, 125, 119, 110, 108, 130, 116, 115, 105, 118,
         115, 112, 137, 106, 120, 117, 119, 111, 106, 121]
M0 = 115
```

```python
import numpy as np

iq_median = np.median(np.array(iq_lst))
iq_median
```

```python
iq_lst.sort()
print(iq_lst)
```

```python
# 중앙값 보다 큰 자료 수 

median_upper = [_ for _ in iq_lst if _ > iq_median]
print(median_upper)
B = len(median_upper)
print(B)
```

```python
# Mo = 115을 제외한 표본 수

n = len([_ for _ in iq_lst if _ !=M0])
n
```

```python
# 표준화 값


Zb = (B- (n/2))/ ((n/4)**0.05)
Zb
```
